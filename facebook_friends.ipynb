```python
friendsfile = open('./friends.json', 'r')
friendsObject = json.load(friendsfile)
friendslist=friendsObject['friends']

print (len(friendslist))
print(friendslist[1])
```

```python
for friend in friendslist:
    print(str(friend['timestamp'])+" = "+friend['name'])
```

```python
import time
from datetime import datetime
for friend in friendslist:
    ts = friend['timestamp']
    print(time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(ts)) +" = "+friend['name'])
    #print(datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S') +" = "+friend['name'])
```

```python
import time
friendyearmonths=[]
for friend in friendslist:
    ts = friend['timestamp']
    #friendyearmonths.append(datetime.utcfromtimestamp(ts).strftime('%Y-%m'))
    friendyearmonths.append(time.strftime('%Y-%m',time.localtime(ts)))
    #+" = "+friend['name'])
#print(friendyearmonths)
#fyms={}
yearmonths=[]
ymfriendcount=[]
lastym="1900-01"
for fym in friendyearmonths:
    if lastym != fym:
        #fyms[fym]=0
        yearmonths.append(fym)
        lastym = fym
print(len(yearmonths))
```

```python
monthlyfriends=[0]*len(yearmonths)
for i in range(len(friendyearmonths)):
    for j in range(len(yearmonths)):
        if yearmonths[j] == friendyearmonths[i]:
            #fyms[yearmonths[j]]=fyms[yearmonths[j]]+1
            monthlyfriends[j]=monthlyfriends[j]+1
            break

print(monthlyfriends)
```
